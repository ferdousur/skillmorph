```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
```

```python
# Upload the Gym data in google drive
drive.mount('/content/drive')
gym = pd.read_csv('/content/drive/MyDrive/datasets/gym_membership.csv')
```

```python
# Display last 5 rows
print(gym.head(5))
```

```python
# Check data types
print(gym.dtypes)
```

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()

# Task 1 Code:
gym_abonoment_count=gym['abonoment_type'].value_counts()
plt.bar(gym_abonoment_count.index, gym_abonoment_count.values, color=['blue', 'green'])
plt.title('Abonoment Type of Standard and Premium Counts')
plt.xlabel('Abonoment Type')
plt.ylabel('Counts')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
# Task 2 Code:
plt.hist(gym['avg_time_in_gym'], bins=20)
plt.title('Average time in Gym')
plt.xlabel('Times')
plt.ylabel('Counts')
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
# # Task 3 Code:
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.title('Age wise visit per week')
plt.xlabel('Age')
plt.ylabel('Visit Per week')
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Gym Gender and Age Statistic')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')

sns.countplot(data=gym, x='days_per_week')
plt.show()
```

```python
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week',)
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
plt.show()
```
